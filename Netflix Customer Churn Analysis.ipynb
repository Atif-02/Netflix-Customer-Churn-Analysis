# Data Visualization

```python
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of Churn',fontsize=20)
plt.show()
```

```python
churn=df['Churn'].value_counts()
plt.figure(figsize=(5,5))
ex=[0,0.1]
plt.pie(churn,labels=churn.index,autopct='%.1f%%',explode=ex,shadow=True)
plt.show()
```

This pie chart will show the 73.5% will not churn and 26.5% will churn

```python
ax=sns.countplot(data=df,x='gender',hue='Churn',palette='viridis')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Churn distribution across genders',fontsize=20)
plt.show()
```

The graph shows that both males and females have similar churn behavior, with most customers not churning.

```python
ax=sns.countplot(data=df,x='SeniorCitizen',hue='Churn',palette='Blues')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Churn distribution across Senior Citizen',fontsize=20)
plt.show()
```

Senior citizens have a higher churn rate compared to non-senior citizens

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x='tenure',bins=72,hue='Churn')
plt.title('Churn distribution accros Tenure',fontsize=22)
plt.show()
```

The graph will show the new customer churn ratio is high according to long-Term(old)

```python
ax=sns.countplot(data=df,x='Contract',hue='Churn',palette='viridis')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Churn distribution across Contract',fontsize=20)
plt.show()
```

The graph will show that the lower-time(month to month) contract coustomer churn ratio is very high accoring to long-term(1year-2year) contract customer.

```python
ax=sns.countplot(data=df,x='PaymentMethod',hue='Churn',palette='Blues')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Churn distribution across PaymentMethod',fontsize=20)
plt.xticks(rotation=30)
plt.show()
```

```python
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

plt.figure(figsize=(18, 12))

for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)
    ax = sns.countplot(data=df, x=col, hue='Churn', palette='Blues', edgecolor='k')
    
    for container in ax.containers:
        ax.bar_label(container, fontsize=9)
    
    plt.title(f'Churn distribution across {col}', fontsize=12)
    plt.xlabel('')
    plt.ylabel('Count')
    plt.xticks(rotation=30)

plt.tight_layout()
plt.suptitle('Churn Distribution Across Service Features', fontsize=20, y=1.02)
plt.show()
```

The charts show that customers who do not subscribe to services like OnlineSecurity, TechSupport, or DeviceProtection have a higher churn rate.
In contrast, customers who use these support-related services are more likely to stay.
Services like StreamingTV and StreamingMovies show no major impact on churn behavior.
